```python
import autogen, groq
from autogen import UserProxyAgent, AssistantAgent

# Define Rent Collection Data
rent_data = {1: 1000, 2: 2000, 3: 3000}  # House numbers with rent
lease_houses = {4, 5, 6}  # Houses with leased rent (automated collection)

# Step 1: Initialize the LLM (Groq API)
llm_config = {
    "model": "gemma2-9b-it",
    "api_key": "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",
    "api_type":"groq"# Replace with your actual API key
}

#  Step 2: Define Rent Collection Agent
rent_collector = AssistantAgent(
    name="RentCollector",
    system_message="You are a rent collection agent. Collect rent from tenants.",
    llm_config=llm_config,
)

#  Step 3: Define Database Agent (stores transactions)
database_agent = AssistantAgent(
    name="DatabaseAgent",
    system_message="You store rent payments and update balances.",
    llm_config=llm_config,
)

#  Step 4: Define the User Agent (simulates tenants)
user_agent = UserProxyAgent(
    name="HouseOwner",
    human_input_mode="NEVER",  # No human intervention
    max_consecutive_auto_reply=2,  # Limits auto-responses
)

#  Step 5: Function to Collect Rent
def collect_rent():
    for house, amount in rent_data.items():
        print(f"\n Collecting rent from House {house}: ${amount}")

        # Simulate interaction
        response = rent_collector.initiate_chat(
            database_agent,
            message=f"Collecting ${amount} from House {house}.",
        )
        print(f"{response['message']}")

    # Process lease payments
    for house in lease_houses:
        print(f"\n House {house} is on lease, rent auto-collected.")

#  Step 6: Start Rent Collection Process
if __name__ == "__main__":
    print("\n Starting Automated Rent Collection System...\n")
    collect_rent()
```

```python
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager

# Rent Prices Dictionary
rent_prices = {1: 1000, 2: 2000, 3: 3000}

# List of tenants and their property types
tenants = {
    "Tenant_A": 1,
    "Tenant_B": 2,
    "Tenant_C": 3,
    "Tenant_D": 4,
    "Tenant_E": 5
}

# LLM Configuration
config_list = [
    {

        'api_key':'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',
        'model':'mixtral-8x7b-32768',
        'api_type': 'groq',
        'max_completion_tokens':2500,# Replace with actual API provider
    }
]

llm_config = {
    #'request_timeout': 600,
    'seed': 42,
    'config_list': config_list,
    'temperature': 0,
}

# Define Agents
task = UserProxyAgent(name='tenant', human_input_mode='NEVER', max_consecutive_auto_reply=3)

collection_agent = AssistantAgent(
    name="collectionagent",
    system_message="Responsible for rent collection notifications.",
    llm_config=llm_config
)

lease_agent = AssistantAgent(
    name="leaseagent",
    system_message="Handles lease agreements for specific property types.",
    llm_config=llm_config
)

finance_agent = AssistantAgent(
    name="finance",
    system_message="Records financial transactions related to rent collection.",
    llm_config=llm_config
)

# Automated Rent Collection Process
def auto_rent_collection():
    for tenant, property_type in tenants.items():
        if property_type in rent_prices:
            rent_amount = rent_prices[property_type]
            print(f"[INFO] Collecting rent for {tenant} (Property Type {property_type}): â‚¹{rent_amount}")
            collection_agent.initiate_chat(task, message=f"{tenant}, your rent for Property Type {property_type} is â‚¹{rent_amount}. Please proceed with payment.")
            #finance_agent.initiate_chat(task, message=f"Rent â‚¹{rent_amount} has been recorded for {tenant}.")
        elif property_type in [4, 5]:
            print(f"[INFO] Lease required for {tenant}. Sending lease document request.")
            lease_agent.initiate_chat(task, message=f"{tenant}, you need to sign a lease document for Property Type {property_type}.")
        else:
            print(f"[ERROR] Invalid property type for {tenant}!")
            task.send_message(f"{tenant}, invalid property type detected. Please enter a valid property number (1-5).")

# Group Chat Setup
group_chat = GroupChat(agents=[task, collection_agent, lease_agent, finance_agent], messages=[])
manager = GroupChatManager(groupchat=group_chat)

# Start Automated Rent Collection
auto_rent_collection()
```

##utility bill agent

```python
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager

# Sample utility bill data (User -> {Utility Type: Amount})
utility_bills = {
    "User_A": {"Electricity": 1500, "Water": 800},
    "User_B": {"Internet": 1200, "Electricity": 2000},
    "User_C": {"Water": 600, "Internet": 1000},
}

# LLM Configuration
config_list = [
    {
        'model': '...',  # Replace with actual model name
        'api_key': '...',  # Replace with actual API key
        'api_type': '...'  # Replace with actual API provider
    }
]

llm_config = {
    #'request_timeout': 600,
    'seed': 42,
    'config_list': config_list,
    'temperature': 0
}

# Define Agents
task = UserProxyAgent(name='user', human_input_mode='NEVER', max_consecutive_auto_reply=3)

billing_agent = AssistantAgent(
    name="billing_agent",
    system_message="Responsible for notifying users about pending utility bills.",
    llm_config=llm_config
)

payment_agent = AssistantAgent(
    name="payment_agent",
    system_message="Handles payment processing and confirms transactions.",
    llm_config=llm_config
)

finance_agent = AssistantAgent(
    name="finance_agent",
    system_message="Records financial transactions related to utility payments.",
    llm_config=llm_config
)

support_agent = AssistantAgent(
    name="support_agent",
    system_message="Assists users with billing disputes or issues.",
    llm_config=llm_config
)

# Automated Bill Payment Process
def auto_bill_payment():
    for user, bills in utility_bills.items():
        for utility, amount in bills.items():
            print(f"[INFO] Processing bill for {user}: {utility} - â‚¹{amount}")
            billing_agent.initiate_chat(task, message=f"{user}, your {utility} bill of â‚¹{amount} is due. Please proceed with payment.")
            payment_agent.initiate_chat(task, message=f"Processing payment of â‚¹{amount} for {user} - {utility}.")
            finance_agent.initiate_chat(task, message=f"Payment of â‚¹{amount} for {utility} recorded for {user}.")

    print("[INFO] All bill payments processed.")

# Group Chat Setup
group_chat = GroupChat(agents=[task, billing_agent, payment_agent, finance_agent, support_agent], message=[])
manager = GroupChatManager(groupchat=group_chat)

# Start Automated Bill Payment
auto_bill_payment()
```
